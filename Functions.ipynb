1. What is the difference between a functions and a method in Python?
  - A function is a standalone block of code that performs a specific task and is defined using the def keyword outside of any class. It can be called independently using its name followed by parentheses, and it does not implicitly receive any arguments unless explicitly passed. On the other hand, a method is essentially a function that is associated with an object and defined within a class. When a method is called, it is invoked on an instance of the class, and it automatically receives the instance as its first argument, conventionally named self. This allows methods to access and manipulate the internal state of the object they belong to. In short, while both functions and methods are defined using the same syntax, methods are bound to objects and have access to their data, whereas functions operate independently.


2. Explain the concept of functions arguments and parameters in Python.
  - In Python, parameters and arguments are closely related concepts used in the context of functions, but they refer to different stages of function usage. Parameters are the variable names listed in a function’s definition—they act as placeholders for the values the function expects to receive. For example, in the function definition def greet(name):, name is a parameter. Arguments, on the other hand, are the actual values passed to the function when it is called. So, when you call greet("Robi"), the string "Robi" is the argument that gets assigned to the parameter name.


3. What are the different ways to define and call a function in Pyton?
 - In Python, functions can be defined and called in various ways to suit different programming needs. The most common way to define a function is using the def keyword followed by the function name and parameters in parentheses. Functions can also include default parameters, allowing them to be called with or without specific arguments. For more flexibility, Python supports variable-length arguments using *args for positional inputs and **kwargs for keyword inputs, enabling functions to handle an arbitrary number of arguments. Additionally, Python allows the creation of anonymous functions using the lambda keyword, which are typically used for short, single-expression functions. Functions can also be nested within other functions, allowing encapsulation and modular design.
When calling functions, Python offers several styles. You can pass arguments positionally, where the order matters, or use keyword arguments to explicitly specify which parameter each value corresponds to. If a function has default parameters, it can be called without providing all arguments. Python also supports argument unpacking, where a tuple or dictionary can be expanded into individual arguments using * or **. Lambda functions are called like regular functions but are often used inline, especially in functional programming contexts. These varied approaches to defining and calling functions make Python highly expressive and adaptable for both simple scripts and complex applications.


4. What is the purpose of the 'return' statement in a Python functions?
  - The return statement in a Python function serves the essential purpose of sending a result back to the caller after the function has completed its task. When a function is executed, it may perform calculations, process data, or carry out operations, and the return statement allows it to output the final value derived from those actions. This returned value can then be stored in a variable, used in expressions, or passed to other functions for further processing. Without a return statement, a function will return None by default, meaning it completes its execution but does not provide any usable output. In essence, return makes a function productive—it transforms it from a passive block of code into an active contributor to the program’s flow and logic.


5. What are iterates in Python and how do the differ from iterables?
  - In Python, iterators and iterables are foundational concepts for working with loops and sequences, but they serve distinct roles. An iterable is any Python object capable of returning its elements one at a time, allowing it to be looped over using a for loop. Common examples of iterables include lists, tuples, strings, dictionaries, and sets. These objects implement the __iter__() method, which returns an iterator.
An iterator, on the other hand, is the object that actually performs the iteration. It keeps track of its current position and knows how to fetch the next item using the __next__() method. Once all items have been returned, calling __next__() again raises a StopIteration exception, signaling that the iteration is complete. You can create an iterator from an iterable using the built-in iter() function.
In short, the iterable is like a book, and the iterator is like a bookmark that helps you read through it page by page. The iterable provides the data, while the iterator provides the mechanism to access that data sequentially.


6. Explain the concept of generators in Python and how they are defined.
 - In Python, generators are a special type of iterator that allow you to iterate over data lazily, meaning they generate values one at a time as needed, rather than storing the entire sequence in memory. This makes generators highly efficient for working with large datasets or infinite sequences, as they conserve memory and improve performance.
Generators are defined using the def keyword like regular functions, but instead of using return, they use the yield keyword to produce a value. When a generator function is called, it doesn’t execute immediately; instead, it returns a generator object. Each time the generator’s __next__() method is called—either explicitly or via a loop—it resumes execution from where it last yielded a value, maintaining its internal state between calls.


7. What are the advantages of using generators over regular functions?
  - Generators offer several compelling advantages over regular functions, especially when dealing with large datasets, streaming data, or complex iteration logic. One of the primary benefits is memory efficiency—generators produce values one at a time using the yield keyword, rather than computing and storing an entire sequence in memory like regular functions with return. This makes them ideal for handling large files, infinite sequences, or data pipelines where loading everything at once would be impractical.
Another advantage is lazy evaluation, meaning generators only compute values when they’re needed. This can significantly improve performance and responsiveness, especially in scenarios where not all results are required immediately. Generators also maintain their internal state between iterations, allowing for clean and elegant code that avoids manual bookkeeping or complex loop logic. Additionally, because generators are inherently iterable, they integrate seamlessly with Python’s for loops and other iteration tools, making them a natural fit for many programming tasks.
In summary, generators provide a powerful, efficient, and readable way to work with sequences and iterative processes, offering advantages in both resource management and code clarity compared to regular functions.


8. What is a lambda function in Python and when is it typically used?
  - A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions created with def, lambda functions are typically used for short, one-time operations where defining a full function would be unnecessary or overly verbose. They consist of a single expression and can take any number of arguments, but they must return the result of that expression directly—no statements or complex logic allowed.
Lambda functions are most commonly used in combination with functions like map(), filter(), and sorted(), where you need to pass a simple function as an argument. They’re also useful in GUI programming, event handling, or any situation where a quick, inline function is needed without cluttering the codebase with named definitions. While lambda functions are concise and convenient, they’re best suited for simple tasks—more complex logic should be handled with regular def functions for clarity and maintainability.


9. Explain the purpose and usage of the 'map()' functions in Python.
  - The map() function in Python is used to apply a given function to every item in an iterable, such as a list or tuple, and returns a new iterable (specifically a map object) containing the results. It is commonly used when you want to transform or process data in a clean and efficient way without writing explicit loops. The syntax is map(function, iterable), where the function is applied to each element of the iterable. This function can be a regular named function or an anonymous lambda function. For example, map(lambda x: x * x, [1, 2, 3]) would return [1, 4, 9]. You can also use map() with multiple iterables if the function accepts multiple arguments, such as map(lambda x, y: x + y, [1, 2], [3, 4]), which would return [4, 6]. The map() function is especially useful in functional programming and data transformation tasks, offering a concise alternative to traditional loops and list comprehensions.


10. What is the difference between 'map()','reduce()', and 'filter()' functions in Python?
  - In Python, map(), reduce(), and filter() are built-in functions that support a functional programming style, but they serve different purposes and operate in distinct ways. The map() function is used to transform each item in an iterable by applying a specified function to it. It returns a new iterable containing the results of the transformation. For example, map(lambda x: x * 2, [1, 2, 3]) would return [2, 4, 6].
The filter() function is used to select items from an iterable based on a condition. It applies a function that returns True or False to each item, and only includes those for which the function returns True. For instance, filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) would return [2, 4].
The reduce() function, which must be imported from the functools module, is used to accumulate values in an iterable into a single result by repeatedly applying a function to pairs of elements. For example, reduce(lambda x, y: x + y, [1, 2, 3, 4]) would return 10.
In summary, map() transforms, filter() selects, and reduce() aggregates. Each plays a unique role in processing data efficiently and concisely.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
from functools import reduce

def sum_even_numbers(numbers):
    # Filter out even numbers
    evens = filter(lambda x: x % 2 == 0, numbers)
    # Sum them using reduce
    return reduce(lambda x, y: x + y, evens, 0)
    print(sum_even_numbers([1, 2, 3, 4, 5, 6]))

```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check and print prime status for numbers from 1 to 200
for num in range(1, 201):
    status = "Prime" if is_prime(num) else "Not Prime"
    print(f"{num}: {status}")
```

5. Create an iterate class in Python that generates the fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage
for power in powers_of_two(10):
    print(power)
```

9. Write a Python program that uses 'map()' to convert a list of temperatures from celcius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Conversion formula: F = C * 9/5 + 32
fahrenheit_temps = list(map(lambda c: c * 9/5 + 32, celsius_temps))

# Print results
print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses 'Filter()' to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, text))

# Example usage
input_string = "Functional programming is powerful!"
result = remove_vowels(input_string)

print("Original:", input_string)
print("Without vowels:", result)
```

11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
   
     Order number       Book title and author        Quantity    Price per item
1.   34587            Learning Python, Mark Lutz       4           40.95
2.   98762           Programming Python, Mark Lutz     5           56.80
3.   77226           Head First Python, Paul Barry     3           32.95
4.   88112           Einfuhrung in Python3, Bern Klien 3           24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10, -E if the value of the order is smaller than 100,00 E.



```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bern Klien", 3, 24.99]
]

def process_orders(order_list):
    result = []
    for order in order_list:
        order_num = order[0]
        quantity = order[2]
        price = order[3]
        total = quantity * price
        if total < 100:
            total += 10  # Add €10 fee
        result.append((order_num, round(total, 2)))
    return result

# Run the routine
processed_orders = process_orders(orders)

# Display results
print(processed_orders)
```

Write a program using Lambda and map.

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bern Klien", 3, 24.99]
]

# Use map with a lambda to compute the required 2-tuples
processed_orders = list(map(
    lambda order: (order[0], round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)),
    orders
))

# Display results
print(processed_orders)
```
